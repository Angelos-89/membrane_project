```python
hh = f.get('RectMesh')
```

```python
hh = np.array(hh)
hh.shape
```

```python
X = np.arange(0, hh.shape[0])
Y = np.arange(0, hh.shape[1])
X, Y = np.meshgrid(X, Y)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, hh, cmap='viridis', linewidth=0, antialiased=False)
plt.xlabel("x")
plt.ylabel("y")
ax.set_zlim(-4, 4)
```

```python
pdf,bin_edges=np.histogram(hh,bins=100,density=True)
```

```python
plt.plot(bin_edges[1:],pdf,'.-')
plt.grid(True)
```
