```python
import h5py
import numpy as np
import pylab
import matplotlib.pyplot as plt
```

```python
f = h5py.File('hfield.h5', 'r')
```

```python
list(f.keys())
```

```python
hh = f.get('RectMesh')
```

```python
hh = np.array(hh)
hh.shape
```

```python
X = np.arange(0, hh.shape[0])
Y = np.arange(0, hh.shape[1])
X, Y = np.meshgrid(X, Y)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, hh, cmap='viridis', linewidth=0, antialiased=False)
plt.xlabel("x")
plt.ylabel("y")
ax.set_zlim(-4, 4)
```
